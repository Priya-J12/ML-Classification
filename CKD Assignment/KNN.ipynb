```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[['classification_yes']]
```

```python
print(independent.shape)
print(dependent.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=1/3, random_state=0)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': [10,50,100],
             'weights':['uniform', 'distance'],
             'algorithm':['auto','ball_tree','kd_tree', 'brute'],
             'metric':['minkowski']}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, refit=True, verbose=3, n_jobs=-1, cv=5, scoring='f1_weighted')#scoring is a evaluation metric
#f1_weighted, provides a performance metric of each class when averaging the F1 scores across different classes
grid.fit(X_train,Y_train.values.ravel()) #values.ravel() is used to remove warning, it flattens multi-dimensional array into a single,1d array.
```

```python
#print best parameter after tuning
print(grid.best_params_)
```

```python
re=grid.cv_results_
#print(re)
grid_predictions=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_predictions)

from sklearn.metrics import classification_report
clf_report = classification_report(Y_test,grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test, grid_predictions, average='weighted')
print("f1_macro value for the best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("Confusion Matrix:\n",cm)
```

```python
print("The Report is:\n",clf_report)
```

```python
print(Y_test.shape)
print(X_test.shape)
```

```python
#Compute Area Under the Receiver Operating Characteristic Curve (ROC AUC) from prediction scores.
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:, 1])
```
